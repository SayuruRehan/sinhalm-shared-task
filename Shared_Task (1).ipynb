````python
# =============================================================================
# 4. IMPROVED CONFIGURATION CLASS
# =============================================================================
class Config:
    """Enhanced configuration with validation and fallback options"""

    # Model settings - using more compatible model first
    model_name = "google/gemma-3-4b-it"  # Start with 2B for compatibility
    # Alternative: "google/gemma-3-4b-it" (can change later)

    # Output directory with timestamp
    timestamp = datetime.now().strftime('%Y%m%d_%H%M')
    output_dir = f"/content/gemma-sinhala-lora-4b-{timestamp}"

    # LoRA settings (optimized for instruction tuning)
    lora_r = 16
    lora_alpha = 32  # Increased for better learning
    lora_dropout = 0.05  # Small dropout for regularization
    target_modules = [
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj"
    ]

    # Training settings (conservative start)
    num_train_epochs = 1
    learning_rate = 2e-4
    per_device_train_batch_size = 1
    gradient_accumulation_steps = 8  # Effective batch size = 8
    max_seq_length = 512

    # Memory optimization
    use_4bit = True
    use_gradient_checkpointing = True
    dataloader_num_workers = 0

    # Logging and saving
    logging_steps = 10
    save_steps = 500
    eval_steps = 500
    save_total_limit = 2

    # Training control
    warmup_steps = 100
    weight_decay = 0.01
    max_grad_norm = 1.0

    # Dataset settings
    use_full_dataset = True
    max_steps = None  # Set to a number for testing (e.g., 100)

    # Validation
    eval_ratio = 0.05  # 5% for validation

    def validate(self):
        """Validate configuration settings"""
        print("🔍 Validating configuration...")

        # Check if output directory can be created
        try:
            os.makedirs(self.output_dir, exist_ok=True)
            print(f"✅ Output directory: {self.output_dir}")
        except Exception as e:
            print(f"❌ Cannot create output directory: {e}")
            return False

        # Validate LoRA settings
        if self.lora_r <= 0 or self.lora_r > 256:
            print(f"❌ Invalid LoRA rank: {self.lora_r}")
            return False

        # Check memory settings
        if torch.cuda.is_available():
            gpu_memory = torch.cuda.get_device_properties(0).total_memory / (1024**3)
            print(f"💾 Available GPU memory: {gpu_memory:.1f} GB")

            if gpu_memory < 8:
                print("⚠️ Low GPU memory detected, reducing batch size")
                self.per_device_train_batch_size = 1
                self.gradient_accumulation_steps = 4
                self.max_seq_length = 256

        print("✅ Configuration validated!")
        return True

    def print_summary(self):
        """Print configuration summary"""
        print("\n📋 Training Configuration:")
        print(f"   🤖 Model: {self.model_name}")
        print(f"   🎯 LoRA Rank: {self.lora_r} (Alpha: {self.lora_alpha})")
        print(f"   📚 Learning Rate: {self.learning_rate}")
        print(f"   🔄 Epochs: {self.num_train_epochs}")
        print(f"   📦 Effective Batch Size: {self.per_device_train_batch_size * self.gradient_accumulation_steps}")
        print(f"   📏 Max Sequence Length: {self.max_seq_length}")
        print(f"   💾 4-bit Quantization: {self.use_4bit}")
        print(f"   🔧 Gradient Checkpointing: {self.use_gradient_checkpointing}")
        print(f"   📂 Output: {self.output_dir}")

# Initialize and validate configuration
config = Config()
if not config.validate():
    print("❌ Configuration validation failed!")
    sys.exit(1)

config.print_summary()

````

````python
# =============================================================================
# 5. MEMORY MANAGEMENT UTILITIES
# =============================================================================
def cleanup_memory():
    """Comprehensive memory cleanup"""
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.cuda.synchronize()
        try:
            torch.cuda.ipc_collect()
        except:
            pass

def print_memory_stats():
    """Print current memory usage"""
    if torch.cuda.is_available():
        for i in range(torch.cuda.device_count()):
            allocated = torch.cuda.memory_allocated(i) / (1024**3)
            reserved = torch.cuda.memory_reserved(i) / (1024**3)
            total = torch.cuda.get_device_properties(i).total_memory / (1024**3)
            print(f"💾 GPU {i}: {allocated:.1f}GB/{total:.1f}GB allocated, {reserved:.1f}GB reserved")

def get_device_info():
    """Get device information"""
    info = {
        "cuda_available": torch.cuda.is_available(),
        "device_count": torch.cuda.device_count() if torch.cuda.is_available() else 0
    }

    if info["cuda_available"]:
        info["device_name"] = torch.cuda.get_device_name(0)
        info["device_memory"] = torch.cuda.get_device_properties(0).total_memory / (1024**3)

    return info

# Print device info
device_info = get_device_info()
print(f"\n🖥️ Device Info:")
print(f"   CUDA Available: {device_info['cuda_available']}")
if device_info['cuda_available']:
    print(f"   Device: {device_info['device_name']}")
    print(f"   Memory: {device_info['device_memory']:.1f} GB")

cleanup_memory()
````

````python
# =============================================================================
# 6. MODEL AND TOKENIZER LOADING WITH FALLBACKS
# =============================================================================
def load_model_and_tokenizer(model_name: str, config: Config):
    """Load model and tokenizer with comprehensive error handling"""
    print(f"\n🔄 Loading model and tokenizer: {model_name}")

    # HuggingFace token
    hf_token = os.getenv("HF_TOKEN") or os.getenv("HUGGINGFACE_TOKEN")

    # 4-bit quantization config
    bnb_config = BitsAndBytesConfig(
        load_in_4bit=config.use_4bit,
        bnb_4bit_quant_type="nf4",
        bnb_4bit_compute_dtype=torch.bfloat16,
        bnb_4bit_use_double_quant=True,
    ) if config.use_4bit else None

    # Load tokenizer
    try:
        print("📝 Loading tokenizer...")
        tokenizer = AutoTokenizer.from_pretrained(
            model_name,
            padding_side="right",
            add_eos_token=True,
            trust_remote_code=True,
            token=hf_token
        )

        # Add pad token if missing
        if tokenizer.pad_token is None:
            tokenizer.pad_token = tokenizer.eos_token
            tokenizer.pad_token_id = tokenizer.eos_token_id

        print(f"✅ Tokenizer loaded - Vocab size: {len(tokenizer)}")

    except Exception as e:
        print(f"❌ Tokenizer loading failed: {e}")
        return None, None

    # Load model with multiple strategies
    model = None
    strategies = [
        {"torch_dtype": torch.bfloat16, "attn_implementation": "eager"},
        {"torch_dtype": torch.float16, "attn_implementation": "eager"},
        {"torch_dtype": torch.float32, "attn_implementation": "eager"},
    ]

    for i, strategy in enumerate(strategies, 1):
        try:
            print(f"🔄 Loading model - Strategy {i}: {strategy}")

            model = AutoModelForCausalLM.from_pretrained(
                model_name,
                quantization_config=bnb_config,
                device_map="auto",
                trust_remote_code=True,
                token=hf_token,
                **strategy
            )

            print(f"✅ Model loaded successfully with strategy {i}")
            break

        except Exception as e:
            print(f"❌ Strategy {i} failed: {str(e)[:100]}...")
            cleanup_memory()
            continue

    if model is None:
        print("❌ All model loading strategies failed")
        return None, None

    # Print model info
    total_params = sum(p.numel() for p in model.parameters())
    print(f"📊 Model info:")
    print(f"   - Total parameters: {total_params:,}")
    print(f"   - Device map: {getattr(model, 'hf_device_map', 'N/A')}")
    print(f"   - Model dtype: {next(model.parameters()).dtype}")

    return model, tokenizer

# Load model and tokenizer
model, tokenizer = load_model_and_tokenizer(config.model_name, config)

if model is None or tokenizer is None:
    print("❌ Failed to load model and tokenizer")
    sys.exit(1)

````

````python
# =============================================================================
# 7. PEFT CONFIGURATION AND MODEL PREPARATION
# =============================================================================
print("\n🔧 Preparing model for LoRA training...")

# Prepare model for k-bit training
model = prepare_model_for_kbit_training(model)

# LoRA configuration
peft_config = LoraConfig(
    r=config.lora_r,
    lora_alpha=config.lora_alpha,
    target_modules=config.target_modules,
    lora_dropout=config.lora_dropout,
    bias="none",
    task_type=TaskType.CAUSAL_LM,
)

# Apply LoRA
model = get_peft_model(model, peft_config)
model.print_trainable_parameters()

# Enable gradient checkpointing
if config.use_gradient_checkpointing:
    model.gradient_checkpointing_enable()
    print("✅ Gradient checkpointing enabled")

print("✅ LoRA configuration applied successfully")
````

````python
# =============================================================================
# 8. DATASET LOADING AND PROCESSING
# =============================================================================
def load_sinhala_dataset():
    """Load Sinhala dataset with error handling"""
    print("\n📊 Loading Sinhala dataset...")

    try:
        # Load the datasets
        print("🔄 Loading flan_zsopt...")
        flan_data = load_dataset("0xAIT/sinhala-flan", "flan_zsopt")

        # Combine all splits
        flan_datasets = []
        for split_name, split_data in flan_data.items():
            flan_datasets.append(split_data)
            print(f"   ✅ {split_name}: {len(split_data):,} examples")

        combined_flan = concatenate_datasets(flan_datasets)
        print(f"✅ Combined flan dataset: {len(combined_flan):,} examples")

        # Try to load t0_zsopt
        try:
            print("🔄 Loading t0_zsopt...")
            t0_data = load_dataset("0xAIT/sinhala-flan", "t0_zsopt")

            t0_datasets = []
            for split_name, split_data in t0_data.items():
                t0_datasets.append(split_data)
                print(f"   ✅ {split_name}: {len(split_data):,} examples")

            combined_t0 = concatenate_datasets(t0_datasets)
            print(f"✅ Combined t0 dataset: {len(combined_t0):,} examples")

            # Combine both
            final_dataset = concatenate_datasets([combined_flan, combined_t0])
            print(f"🎯 Final combined dataset: {len(final_dataset):,} examples")

        except Exception as e:
            print(f"⚠️ t0_zsopt loading failed: {e}")
            print("📊 Using only flan_zsopt dataset")
            final_dataset = combined_flan

        return final_dataset

    except Exception as e:
        print(f"❌ Dataset loading failed: {e}")
        return None

def create_sample_dataset(size: int = 1000):
    """Create sample dataset for testing"""
    print(f"\n🧪 Creating sample dataset with {size} examples...")

    sample_data = [
        {"text": "### Instruction:\nසිංහල භාෂාවේ ප්‍රධාන ලක්ෂණ මොනවාද?\n\n### Response:\nසිංහල භාෂාවේ ප්‍රධාන ලක්ෂණ වන්නේ එහි අක්ෂර මාලාව, ව්‍යාකරණ සහ සංස්කෘතික වටිනාකම් ය."},
        {"text": "### Instruction:\nකොළඹ නගරය ගැන කියන්න.\n\n### Response:\nකොළඹ ශ්‍රී ලංකාවේ වාණිජ අගනුවර වන අතර එය ප්‍රධාන වරාය නගරයකි."},
        {"text": "### Instruction:\nශ්‍රී ලංකාවේ ඉතිහාසය ගැන කියන්න.\n\n### Response:\nශ්‍රී ලංකාවේ ඉතිහාසය වසර දහස් ගණනක් දක්වා දිව යයි."},
        {"text": "### Instruction:\nසිංහල කවි පරම්පරාව ගැන කියන්න.\n\n### Response:\nසිංහල කවි පරම්පරාව ඉතා පුරාණ හා සුවිශේෂී වේ."},
        {"text": "### Instruction:\nබුදු දහම ගැන කියන්න.\n\n### Response:\nබුදු දහම ශ්‍රී ලංකාවේ ප්‍රධාන ආගම වන අතර එය සිහල් සංස්කෘතියට ගැඹුරු බලපෑමක් කර ඇත."}
    ]

    # Repeat to create desired size
    multiplier = (size // len(sample_data)) + 1
    expanded_data = (sample_data * multiplier)[:size]

    return Dataset.from_list(expanded_data)

def format_instruction_data(example):
    """Format data for instruction tuning"""
    # Handle different column name variations
    instruction = (
        example.get('Translated Input', '') or
        example.get('Original Input', '') or
        example.get('instruction', '') or
        str(example.get('input', ''))
    ).strip()

    response = (
        example.get('Translated Target', '') or
        example.get('Original Target', '') or
        example.get('response', '') or
        str(example.get('target', ''))
    ).strip()

    if instruction and response:
        formatted_text = f"### Instruction:\n{instruction}\n\n### Response:\n{response}"
        return {"text": formatted_text}

    return {"text": ""}

# Load and process dataset
if config.use_full_dataset:
    dataset = load_sinhala_dataset()

    if dataset is not None:
        print(f"\n📋 Dataset info:")
        print(f"   Columns: {dataset.column_names}")
        print(f"   Total examples: {len(dataset):,}")

        # Show sample
        sample = dataset[0]
        print(f"\n🔍 Sample data:")
        for key, value in sample.items():
            print(f"   {key}: {str(value)[:100]}...")

        # Format dataset
        print("\n🔄 Formatting dataset...")
        formatted_dataset = dataset.map(
            format_instruction_data,
            remove_columns=dataset.column_names,
            desc="Formatting dataset"
        )

        # Filter empty examples
        formatted_dataset = formatted_dataset.filter(
            lambda x: len(x["text"].strip()) > 10,
            desc="Filtering empty examples"
        )

        print(f"✅ Formatted dataset: {len(formatted_dataset):,} examples")

        # Show formatted sample
        if len(formatted_dataset) > 0:
            print(f"\n📝 Formatted sample:")
            sample_text = formatted_dataset[0]['text']
            print(sample_text[:300] + "..." if len(sample_text) > 300 else sample_text)
    else:
        print("⚠️ Dataset loading failed, using sample data")
        formatted_dataset = create_sample_dataset()
else:
    print("🧪 Using sample dataset for testing")
    formatted_dataset = create_sample_dataset()

# Create train/validation split
print(f"\n✂️ Creating train/validation split...")
dataset_split = formatted_dataset.train_test_split(
    test_size=config.eval_ratio,
    seed=42,
    shuffle=True
)

train_dataset = dataset_split['train']
eval_dataset = dataset_split['test']

print(f"🚂 Training examples: {len(train_dataset):,}")
print(f"🔍 Validation examples: {len(eval_dataset):,}")
````

````python
# =============================================================================
# 9. TOKENIZATION AND DATA PREPARATION
# =============================================================================
def tokenize_function(examples):
    """Tokenize examples for training"""
    tokenized = tokenizer(
        examples["text"],
        truncation=True,
        padding="max_length",
        max_length=config.max_seq_length,
        return_tensors=None,
    )

    # Set labels for causal LM (copy input_ids)
    tokenized["labels"] = tokenized["input_ids"].copy()

    return tokenized

print("\n🔄 Tokenizing datasets...")

# Tokenize train dataset
train_dataset = train_dataset.map(
    tokenize_function,
    batched=True,
    remove_columns=train_dataset.column_names,
    desc="Tokenizing training data"
)

# Tokenize eval dataset
eval_dataset = eval_dataset.map(
    tokenize_function,
    batched=True,
    remove_columns=eval_dataset.column_names,
    desc="Tokenizing validation data"
)

print(f"✅ Tokenization complete!")
print(f"   📊 Train samples: {len(train_dataset):,}")
print(f"   📊 Eval samples: {len(eval_dataset):,}")

# Verify tokenization
sample = train_dataset[0]
print(f"🔍 Tokenization check:")
print(f"   - Input IDs length: {len(sample['input_ids'])}")
print(f"   - Labels length: {len(sample['labels'])}")

````

````python
# =============================================================================
# 10. TRAINING SETUP AND EXECUTION
# =============================================================================
# Data collator
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False,
)

# Training arguments
training_args = TrainingArguments(
    output_dir=config.output_dir,
    num_train_epochs=config.num_train_epochs,
    per_device_train_batch_size=config.per_device_train_batch_size,
    per_device_eval_batch_size=config.per_device_train_batch_size,
    gradient_accumulation_steps=config.gradient_accumulation_steps,
    learning_rate=config.learning_rate,
    weight_decay=config.weight_decay,
    warmup_steps=config.warmup_steps,
    max_grad_norm=config.max_grad_norm,

    # Steps configuration
    max_steps=config.max_steps if config.max_steps else -1,

    # Logging and evaluation
    logging_steps=config.logging_steps,
    eval_steps=config.eval_steps,
    eval_strategy="steps",
    save_steps=config.save_steps,
    save_total_limit=config.save_total_limit,

    # Optimization
    bf16=True,
    dataloader_num_workers=config.dataloader_num_workers,
    remove_unused_columns=False,
    gradient_checkpointing=config.use_gradient_checkpointing,

    # Disable reporting
    report_to="none",

    # Save settings
    load_best_model_at_end=False,
    save_safetensors=True,
    ddp_find_unused_parameters=False,
)

print(f"✅ Training arguments configured")

# Initialize trainer
print(f"\n🔄 Initializing trainer...")
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
)

# Calculate training estimates
if config.max_steps:
    total_steps = config.max_steps
else:
    steps_per_epoch = len(train_dataset) // (
        training_args.per_device_train_batch_size *
        training_args.gradient_accumulation_steps
    )
    total_steps = steps_per_epoch * training_args.num_train_epochs

print(f"\n⏱️ Training estimates:")
print(f"   📊 Total steps: {total_steps:,}")
print(f"   🕐 Estimated time: {total_steps * 0.02:.1f} hours")

# Pre-training memory check
print(f"\n📊 Pre-training memory:")
print_memory_stats()

print(f"\n🚀 Starting training...")
print(f"⏰ Start time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

# Training with comprehensive error handling
try:
    trainer.train()

    print("✅ Training completed successfully!")

    # Save model
    final_model_path = f"{config.output_dir}/final_model"
    print(f"💾 Saving model to: {final_model_path}")

    trainer.save_model(final_model_path)
    tokenizer.save_pretrained(final_model_path)

    print("✅ Model saved successfully!")

except KeyboardInterrupt:
    print("⏹️ Training interrupted by user")

except RuntimeError as e:
    if "out of memory" in str(e).lower():
        print(f"❌ Out of Memory Error: {e}")
        print("💡 Suggestions:")
        print("   - Reduce batch_size or max_seq_length")
        print("   - Increase gradient_accumulation_steps")
        print("   - Reduce lora_r")
        print(f"   - Current: batch={config.per_device_train_batch_size}, seq_len={config.max_seq_length}, lora_r={config.lora_r}")
    else:
        print(f"❌ Runtime Error: {e}")

except Exception as e:
    print(f"❌ Training failed: {e}")
    import traceback
    traceback.print_exc()

finally:
    cleanup_memory()

print(f"\n⏰ Completed at: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

# Final summary
print(f"\n🎯 Training Summary:")
print(f"   🤖 Model: {config.model_name}")
print(f"   📊 Training samples: {len(train_dataset):,}")
print(f"   🎯 LoRA rank: {config.lora_r}")
print(f"   💾 Output: {config.output_dir}")
print("✅ Training complete!")
````

````python
# =============================================================================
# POST-TRAINING TESTING SCRIPT FOR SINHALA GEMMA LORA
# Specifically designed for your trained model
# =============================================================================

import torch
import os
import json
import time
import gc
from datetime import datetime
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    BitsAndBytesConfig
)
from peft import PeftModel

# =============================================================================
# 1. EXACT MODEL CONFIGURATION FROM YOUR TRAINING
# =============================================================================
# Your exact model paths from the training output
BASE_MODEL_NAME = "google/gemma-3-4b-it"
LORA_MODEL_PATH = "/content/gemma-sinhala-lora-4b-20250803_1228/final_model"
OUTPUT_DIR = "/content/test_results_sinhala_gemma"

# Test configuration
class TestConfig:
    # Generation parameters optimized for Sinhala
    max_new_tokens = 512
    temperature = 0.7
    top_p = 0.9
    top_k = 50
    repetition_penalty = 1.1
    do_sample = True

    # Device settings
    device = "cuda" if torch.cuda.is_available() else "cpu"

    # Output settings
    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')

config = TestConfig()

def cleanup_memory():
    """Clean GPU memory"""
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.cuda.synchronize()
    gc.collect()

def print_gpu_memory():
    """Print GPU memory usage"""
    if torch.cuda.is_available():
        allocated = torch.cuda.memory_allocated() / 1024**3
        reserved = torch.cuda.memory_reserved() / 1024**3
        total = torch.cuda.get_device_properties(0).total_memory / 1024**3
        print(f"💾 GPU: {allocated:.1f}GB allocated, {reserved:.1f}GB reserved, {total:.1f}GB total")

# =============================================================================
# 2. MODEL LOADING (MATCHING YOUR TRAINING SETUP)
# =============================================================================
print("🚀 SINHALA GEMMA LORA MODEL TESTING")
print("=" * 50)
print(f"📅 Test Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"🤖 Base Model: {BASE_MODEL_NAME}")
print(f"🎯 LoRA Model: {LORA_MODEL_PATH}")

# Create output directory
os.makedirs(OUTPUT_DIR, exist_ok=True)
print(f"📁 Output Directory: {OUTPUT_DIR}")

print("\n🔄 Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL_NAME)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
print("✅ Tokenizer loaded")

print("\n🔄 Loading base model with 4-bit quantization...")
# Use same quantization as your training
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

base_model = AutoModelForCausalLM.from_pretrained(
    BASE_MODEL_NAME,
    quantization_config=bnb_config,
    device_map="auto",
    torch_dtype=torch.bfloat16,
    trust_remote_code=True
)
print("✅ Base model loaded")

print("\n🔄 Loading your trained LoRA weights...")
try:
    model = PeftModel.from_pretrained(base_model, LORA_MODEL_PATH)
    print("✅ LoRA weights loaded successfully!")
    lora_loaded = True
    model_type = "Fine-tuned LoRA Model"
except Exception as e:
    print(f"❌ Could not load LoRA weights: {e}")
    print("📝 Falling back to base model...")
    model = base_model
    lora_loaded = False
    model_type = "Base Model (No LoRA)"

model.eval()
print(f"🧠 Using: {model_type}")
print_gpu_memory()

# =============================================================================
# 3. YOUR SPECIFIC SINHALA TEST CASES
# =============================================================================
sinhala_test_cases = [
    {
        "id": "human_evolution",
        "prompt": "මානව පරිණාම‍ය ගැන වාක්‍ය 5ක ඡේද‍යක් ලි‍යන්න.",
        "description": "Generate 5-sentence paragraph about human evolution in Sinhala",
        "expected_keywords": ["මානව", "පරිණාම", "විකාශන"]
    },
    {
        "id": "kandy_history_comprehension",
        "prompt": """පහත විස්තර‍ය කි‍යවා පිළිතුරු ලබා දෙන්න.
විස්තර‍ය: "ක්‍රි.ව. 1592 වර්ෂයේ දී සෙංකඩගල අග නගරය බවට පත් විය. ක්‍රි.ව. 16, 17, 18 යන ශතවර්ෂවල දී ඇති වූ පෘතුගීසි, ලන්දේසි ආක්‍රමණ ද අවසාන ලෙස බ්‍රිතාන්‍යය ආක්‍රමණයන් ද සාර්ථකව මැඩපවත්වන ලදී. මහනුවර 19 වන ශත වර්ෂය තෙක් ස්වාධීන රාජ්‍යයක් ලෙස පැවතිනි. ඉන් පසු බ්‍රිතාන්‍යයන් සමඟ ඇති වූ දෙවන මහනුවර සංග්‍රාමයේ දී බ්‍රිතාන්‍යයන් 1815 පෙබරවාරි මස 15 වන දින නගරය වෙත ළඟාවෙන ලදි. 1815 මාර්තු 2 වන දින උඩරට ගිවිසුම මගින් මහනුවර ඇතුලු සමස්ථ දිවයින ම බ්‍රිතාන්‍යයන්ගේ යටත් විජිතයක් බවට පත් විය.එසේ වීමට හේතුව වුයේ උඩරට රදළ වරුන්ගේ ක්‍රියා කලාපයයි. ඔවුන් එකල රජ කල ශ්‍රී වික්‍රම රාජසිංහ රජු මරා රජ වීමට උස්සහ කලහ. මෙම කත් එදිල්ලේ ප්‍රතිඵලයක් ලෙස උඩරට යටත් විය."
ප්‍රශ්න‍ය: මහනුවර කීවෙනි සි‍යවස තෙක් ස්වාධීනව පැවැති‍යේ ද?""",
        "description": "Reading comprehension about Kandy history",
        "expected_answer": "19 වන ශත වර්ෂය",
        "expected_keywords": ["19", "ශත", "වර්ෂය"]
    },
    {
        "id": "words_starting_with_a",
        "prompt": "අ‍'යන්නෙන් ආරම්භ වන වචන 5ක් කි‍යන්න",
        "description": "Generate 5 Sinhala words starting with 'අ'",
        "expected_pattern": "words starting with අ"
    },
    {
        "id": "simple_story",
        "prompt": "හාල් මැස්සාගේ කෙටි කතාවක් ලියන්න.",
        "description": "Write a short story about a chameleon",
        "expected_keywords": ["හාල්", "මැස්සා", "කතාව"]
    },
    {
        "id": "math_explanation",
        "prompt": "2 + 3 = 5 මේ ගණිත ගණනය සිංහලෙන් පැහැදිලි කරන්න.",
        "description": "Explain simple addition in Sinhala",
        "expected_keywords": ["දෙක", "තුන", "පහ", "එකතු"]
    }
]

# =============================================================================
# 4. GENERATION FUNCTION (OPTIMIZED FOR GEMMA)
# =============================================================================
def generate_sinhala_response(prompt, max_tokens=512):
    """Generate response using your trained model"""

    # Format prompt using Gemma's chat template
    formatted_prompt = f"<start_of_turn>user\n{prompt}<end_of_turn>\n<start_of_turn>model\n"

    # Tokenize input
    inputs = tokenizer(
        formatted_prompt,
        return_tensors="pt",
        truncation=True,
        max_length=1024,  # Give enough context
        padding=False
    ).to(config.device)

    start_time = time.time()

    # Generate response
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_tokens,
            temperature=config.temperature,
            top_p=config.top_p,
            top_k=config.top_k,
            repetition_penalty=config.repetition_penalty,
            do_sample=config.do_sample,
            pad_token_id=tokenizer.eos_token_id,
            eos_token_id=tokenizer.eos_token_id,
        )

    generation_time = time.time() - start_time

    # Decode and clean response
    full_response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract only the model's response
    if "<start_of_turn>model\n" in full_response:
        response = full_response.split("<start_of_turn>model\n")[-1].strip()
    else:
        response = full_response.strip()

    return response, generation_time

# =============================================================================
# 5. EXECUTE TESTS
# =============================================================================
print("\n🧪 STARTING COMPREHENSIVE SINHALA TESTING")
print("=" * 50)

test_results = []
total_start_time = time.time()

for i, test_case in enumerate(sinhala_test_cases, 1):
    print(f"\n📝 Test {i}/{len(sinhala_test_cases)}: {test_case['id'].upper()}")
    print(f"📄 {test_case['description']}")
    print(f"❓ Prompt: {test_case['prompt'][:80]}{'...' if len(test_case['prompt']) > 80 else ''}")

    try:
        # Generate response
        response, gen_time = generate_sinhala_response(test_case['prompt'])

        # Analyze response
        analysis = {
            "length": len(response),
            "word_count": len(response.split()),
            "contains_sinhala": any(ord(char) >= 0x0D80 and ord(char) <= 0x0DFF for char in response),
            "generation_time": gen_time
        }

        # Check for expected content
        if "expected_keywords" in test_case:
            found_keywords = [kw for kw in test_case["expected_keywords"] if kw in response]
            analysis["keywords_found"] = found_keywords
            analysis["keyword_score"] = len(found_keywords) / len(test_case["expected_keywords"])

        if "expected_answer" in test_case:
            analysis["correct_answer"] = test_case["expected_answer"] in response

        if test_case["id"] == "words_starting_with_a":
            words_with_a = [word for word in response.split() if word.startswith('අ')]
            analysis["words_starting_with_a"] = words_with_a
            analysis["word_count_with_a"] = len(words_with_a)

        # Store result
        result = {
            "test_id": test_case["id"],
            "prompt": test_case["prompt"],
            "response": response,
            "analysis": analysis,
            "success": True,
            "timestamp": datetime.now().isoformat()
        }

        test_results.append(result)

        # Print results
        print(f"✅ Generated in {gen_time:.2f}s")
        print(f"📊 Length: {analysis['length']} chars, {analysis['word_count']} words")
        print(f"🔤 Contains Sinhala: {'✅' if analysis['contains_sinhala'] else '❌'}")

        # Print response preview
        preview = response[:200] + "..." if len(response) > 200 else response
        print(f"💬 Response: {preview}")

        # Specific evaluations
        if test_case["id"] == "kandy_history_comprehension":
            if "19" in response:
                print("🎯 CORRECT: Found '19' in response!")
            else:
                print("⚠️ Expected '19 වන ශත වර්ෂය' not clearly found")

        elif test_case["id"] == "words_starting_with_a":
            if analysis.get("word_count_with_a", 0) > 0:
                print(f"🎯 Found {analysis['word_count_with_a']} words starting with 'අ': {', '.join(analysis['words_starting_with_a'][:3])}")
            else:
                print("⚠️ No words starting with 'අ' found")

        elif "keywords_found" in analysis:
            if analysis["keyword_score"] > 0:
                print(f"🎯 Keywords found: {', '.join(analysis['keywords_found'])}")
            else:
                print("⚠️ No expected keywords found")

    except Exception as e:
        print(f"❌ ERROR: {e}")
        result = {
            "test_id": test_case["id"],
            "prompt": test_case["prompt"],
            "response": f"ERROR: {str(e)}",
            "success": False,
            "error": str(e),
            "timestamp": datetime.now().isoformat()
        }
        test_results.append(result)

    # Clean memory after each test
    cleanup_memory()

total_test_time = time.time() - total_start_time

# =============================================================================
# 6. RESULTS SUMMARY AND SAVING
# =============================================================================
print("\n" + "=" * 50)
print("📊 TESTING RESULTS SUMMARY")
print("=" * 50)

successful_tests = [r for r in test_results if r["success"]]
failed_tests = [r for r in test_results if not r["success"]]

print(f"✅ Successful Tests: {len(successful_tests)}/{len(test_results)}")
print(f"❌ Failed Tests: {len(failed_tests)}")
print(f"⏱️ Total Testing Time: {total_test_time:.2f} seconds")

if successful_tests:
    avg_gen_time = sum(r["analysis"]["generation_time"] for r in successful_tests) / len(successful_tests)
    print(f"📈 Average Generation Time: {avg_gen_time:.2f} seconds")

print(f"🧠 Model Used: {model_type}")
print(f"🎯 LoRA Status: {'✅ Loaded' if lora_loaded else '❌ Not Loaded'}")

# Save detailed results
results_file = os.path.join(OUTPUT_DIR, f"sinhala_test_results_{config.timestamp}.json")
with open(results_file, 'w', encoding='utf-8') as f:
    json.dump({
        "test_info": {
            "base_model": BASE_MODEL_NAME,
            "lora_path": LORA_MODEL_PATH,
            "lora_loaded": lora_loaded,
            "model_type": model_type,
            "test_date": datetime.now().isoformat(),
            "total_tests": len(test_results),
            "successful_tests": len(successful_tests),
            "total_time": total_test_time
        },
        "results": test_results
    }, f, ensure_ascii=False, indent=2)

print(f"💾 Detailed results saved: {results_file}")

# Save human-readable summary
summary_file = os.path.join(OUTPUT_DIR, f"test_summary_{config.timestamp}.txt")
with open(summary_file, 'w', encoding='utf-8') as f:
    f.write("SINHALA GEMMA LORA MODEL - TEST SUMMARY\n")
    f.write("=" * 45 + "\n\n")
    f.write(f"Test Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
    f.write(f"Base Model: {BASE_MODEL_NAME}\n")
    f.write(f"LoRA Model: {LORA_MODEL_PATH}\n")
    f.write(f"LoRA Loaded: {'Yes' if lora_loaded else 'No'}\n")
    f.write(f"Training Time: 1.84 hours\n")
    f.write(f"Training Samples: 50,000\n")
    f.write(f"LoRA Rank: 16\n\n")

    f.write("TEST RESULTS:\n")
    f.write("-" * 20 + "\n")

    for result in test_results:
        f.write(f"\nTEST: {result['test_id'].upper()}\n")
        f.write(f"Success: {'✅' if result['success'] else '❌'}\n")
        if result['success']:
            f.write(f"Generation Time: {result['analysis']['generation_time']:.2f}s\n")
            f.write(f"Response Length: {result['analysis']['length']} chars\n")
        f.write(f"Prompt: {result['prompt'][:100]}...\n")
        f.write(f"Response: {result['response'][:300]}...\n")
        f.write("-" * 40 + "\n")

print(f"📄 Summary saved: {summary_file}")

# =============================================================================
# 7. SPECIFIC PERFORMANCE EVALUATION
# =============================================================================
print(f"\n🎯 SPECIFIC PERFORMANCE EVALUATION:")
print("=" * 40)

# Check human evolution test
evolution_test = next((r for r in successful_tests if r["test_id"] == "human_evolution"), None)
if evolution_test:
    response = evolution_test["response"]
    sentences = [s.strip() for s in response.split('.') if s.strip()]
    print(f"📝 Human Evolution:")
    print(f"   Sentences: {len(sentences)}")
    print(f"   Contains 'මානව': {'✅' if 'මානව' in response else '❌'}")
    print(f"   Contains 'පරිණාම': {'✅' if 'පරිණාම' in response else '❌'}")

# Check reading comprehension
kandy_test = next((r for r in successful_tests if r["test_id"] == "kandy_history_comprehension"), None)
if kandy_test:
    correct = kandy_test["analysis"].get("correct_answer", False)
    print(f"📚 Reading Comprehension:")
    print(f"   Correct Answer: {'✅ YES' if correct else '❌ NO'}")
    print(f"   Contains '19': {'✅' if '19' in kandy_test['response'] else '❌'}")

# Check word generation
words_test = next((r for r in successful_tests if r["test_id"] == "words_starting_with_a"), None)
if words_test:
    word_count = words_test["analysis"].get("word_count_with_a", 0)
    print(f"📝 Word Generation:")
    print(f"   Words with 'අ': {word_count}")
    if word_count > 0:
        words = words_test["analysis"].get("words_starting_with_a", [])
        print(f"   Examples: {', '.join(words[:5])}")

print(f"\n🎉 TESTING COMPLETED SUCCESSFULLY!")
print(f"📁 All results saved in: {OUTPUT_DIR}")

# Final cleanup
cleanup_memory()
print_gpu_memory()
print("✅ Testing script finished!")
````

````python
# =============================================================================
# COMPLETE SINHALA TESTING SUITE WITH EASY QUESTIONS
# Comprehensive test suite for your trained Sinhala Gemma model
# =============================================================================

import torch
import os
import json
import time
import gc
from datetime import datetime
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import PeftModel

# Model configuration
BASE_MODEL_NAME = "google/gemma-3-4b-it"
LORA_MODEL_PATH = "/content/gemma-sinhala-lora-4b-20250803_1228/final_model"
OUTPUT_DIR = "/content/complete_sinhala_test_results"

# =============================================================================
# ALL TEST CASES - FROM EASY TO ADVANCED
# =============================================================================

# Easy basic vocabulary tests
easy_tests = [
    {
        "id": "colors_sinhala",
        "prompt": "සිංහල වර්ණ 5ක් නම් කරන්න.",
        "description": "Name 5 colors in Sinhala",
        "expected_keywords": ["රතු", "නිල්", "කහ", "කොළ", "සුදු"],
        "difficulty": "easy",
        "category": "vocabulary"
    },
    {
        "id": "numbers_1_to_10",
        "prompt": "1 සිට 10 දක්වා සංඛ්‍යා සිංහලෙන් ලියන්න.",
        "description": "Write numbers 1-10 in Sinhala",
        "expected_keywords": ["එක", "දෙක", "තුන", "හතර", "පහ"],
        "difficulty": "easy",
        "category": "vocabulary"
    },
    {
        "id": "family_members",
        "prompt": "පවුලේ සාමාජිකයන්ගේ නම් 5ක් කියන්න.",
        "description": "Name 5 family members in Sinhala",
        "expected_keywords": ["මව", "පිය", "සහෝදර", "සහෝදරි", "ආච්චි"],
        "difficulty": "easy",
        "category": "vocabulary"
    },
    {
        "id": "days_of_week",
        "prompt": "සතියේ දින නම් කරන්න.",
        "description": "Name days of the week",
        "expected_keywords": ["සඳුදා", "අඟහරුවාදා", "බදාදා", "සිකුරාදා"],
        "difficulty": "easy",
        "category": "vocabulary"
    },
    {
        "id": "body_parts",
        "prompt": "ශරීරයේ කොටස් 6ක් නම් කරන්න.",
        "description": "Name 6 body parts",
        "expected_keywords": ["ඔළුව", "ඇස", "කන", "අත", "පාද"],
        "difficulty": "easy",
        "category": "vocabulary"
    }
]

# Easy conversation tests
conversation_tests = [
    {
        "id": "greetings",
        "prompt": "සුභ දවසක්! ඔබට කෙසේද?",
        "description": "Respond to a greeting",
        "expected_keywords": ["සුභ", "හොඳින්", "ස්තූතිය"],
        "difficulty": "easy",
        "category": "conversation"
    },
    {
        "id": "self_introduction",
        "prompt": "ඔබගේ නම කියා ඔබ ගැන කෙටියෙන් කියන්න.",
        "description": "Brief self introduction",
        "expected_keywords": ["නම", "මම", "කරනවා"],
        "difficulty": "easy",
        "category": "conversation"
    },
    {
        "id": "weather_talk",
        "prompt": "අද කාලගුණය ගැන වාක්‍ය 3ක් ලියන්න.",
        "description": "Write about weather",
        "expected_keywords": ["අද", "කාලගුණය", "හිරු", "වැස්ස"],
        "difficulty": "easy",
        "category": "conversation"
    }
]

# Easy reading comprehension
reading_tests = [
    {
        "id": "short_story_reading",
        "prompt": """පහත කතාව කියවා ප්‍රශ්නයට පිළිතුරු දෙන්න.
කතාව: "මල්ලි ඊයේ පාසලට ගියා. ඔහු ගණිත පාඩම ඉගෙන ගත්තා. ගුරුතුමා ඔහුට ප්‍රශංසා කළා."
ප්‍රශ්නය: මල්ලි කුමක් ඉගෙන ගත්තේද?""",
        "description": "Simple story comprehension",
        "expected_answer": "ගණිත",
        "expected_keywords": ["ගණිත", "පාඩම"],
        "difficulty": "easy",
        "category": "comprehension"
    },
    {
        "id": "animal_facts",
        "prompt": """පහත තොරතුර කියවා ප්‍රශ්නයට පිළිතුරු දෙන්න.
තොරතුර: "අලි ගෙම මහත්ම භූමිෆල සතුන්ය. ඔවුන් දිනකට කිලෝ 300ක් පමණ කොළ කනවා."
ප්‍රශ්නය: අලි ගෙම දිනකට කොළ කීයක් කනවාද?""",
        "description": "Animal facts comprehension",
        "expected_answer": "කිලෝ 300",
        "expected_keywords": ["300", "කිලෝ"],
        "difficulty": "easy",
        "category": "comprehension"
    }
]

# Easy creative writing
creative_tests = [
    {
        "id": "my_school",
        "prompt": "මගේ පාසල ගැන වාක්‍ය 4ක් ලියන්න.",
        "description": "Write about school",
        "expected_keywords": ["පාසල", "ගුරුතුමා", "මිතුරන්", "ඉගෙන"],
        "difficulty": "easy",
        "category": "creative"
    },
    {
        "id": "favorite_food",
        "prompt": "ඔබගේ ප්‍රියතම ආහාරය ගැන කෙටි විස්තරයක් ලියන්න.",
        "description": "Describe favorite food",
        "expected_keywords": ["ආහාර", "ප්‍රිය", "රස"],
        "difficulty": "easy",
        "category": "creative"
    }
]

# Easy math tests
math_tests = [
    {
        "id": "basic_addition",
        "prompt": "5 + 7 = ? මේ ගණනය සිංහලෙන් කරලා පිළිතුර දෙන්න.",
        "description": "Simple addition",
        "expected_answer": "12",
        "expected_keywords": ["පහ", "හත", "දොළහ"],
        "difficulty": "easy",
        "category": "math"
    },
    {
        "id": "basic_subtraction",
        "prompt": "10 - 4 = ? මේ ගණනය සිංහලෙන් පැහැදිලි කරන්න.",
        "description": "Simple subtraction",
        "expected_answer": "6",
        "expected_keywords": ["දහය", "හතර", "හය"],
        "difficulty": "easy",
        "category": "math"
    }
]

# Easy translation tests
translation_tests = [
    {
        "id": "english_to_sinhala_basic",
        "prompt": "\"I love my mother\" මේ වාක්‍යය සිංහලට පරිවර්තනය කරන්න.",
        "description": "Basic English to Sinhala",
        "expected_keywords": ["මම", "ආදරය", "අම්මා"],
        "difficulty": "easy",
        "category": "translation"
    }
]

# Easy cultural knowledge
cultural_tests = [
    {
        "id": "sri_lankan_cities",
        "prompt": "ශ්‍රී ලංකාවේ ප්‍රධාන නගර 5ක් නම් කරන්න.",
        "description": "Name major Sri Lankan cities",
        "expected_keywords": ["කොළඹ", "කණ්ඩි", "ගාල්ල", "ජාෆ්නා"],
        "difficulty": "easy",
        "category": "culture"
    },
    {
        "id": "sinhala_new_year",
        "prompt": "සිංහල අවුරුදු ගැන වාක්‍ය 3ක් ලියන්න.",
        "description": "Write about Sinhala New Year",
        "expected_keywords": ["අවුරුදු", "උත්සව", "කිරිබත්"],
        "difficulty": "easy",
        "category": "culture"
    }
]

# Your original advanced test cases
advanced_tests = [
    {
        "id": "human_evolution",
        "prompt": "මානව පරිණාම‍ය ගැන වාක්‍ය 5ක ඡේද‍යක් ලි‍යන්න.",
        "description": "Human evolution paragraph",
        "expected_keywords": ["මානව", "පරිණාම", "විකාශන"],
        "difficulty": "advanced",
        "category": "knowledge"
    },
    {
        "id": "kandy_history_comprehension",
        "prompt": """පහත විස්තර‍ය කි‍යවා පිළිතුරු ලබා දෙන්න.
විස්තර‍ය: "ක්‍රි.ව. 1592 වර්ෂයේ දී සෙංකඩගල අග නගරය බවට පත් විය. මහනුවර 19 වන ශත වර්ෂය තෙක් ස්වාධීන රාජ්‍යයක් ලෙස පැවතිනි."
ප්‍රශ්න‍ය: මහනුවර කීවෙනි සි‍යවස තෙක් ස්වාධීනව පැවැති‍යේ ද?""",
        "description": "Historical reading comprehension",
        "expected_answer": "19 වන ශත වර්ෂය",
        "expected_keywords": ["19", "ශත"],
        "difficulty": "advanced",
        "category": "comprehension"
    },
    {
        "id": "words_starting_with_a",
        "prompt": "අ‍'යන්නෙන් ආරම්භ වන වචන 5ක් කි‍යන්න",
        "description": "Words starting with 'අ'",
        "difficulty": "medium",
        "category": "vocabulary"
    }
]

# Combine all tests
all_test_cases = (
    easy_tests +
    conversation_tests +
    reading_tests +
    creative_tests +
    math_tests +
    translation_tests +
    cultural_tests +
    advanced_tests
)

# =============================================================================
# TESTING CONFIGURATION AND SETUP
# =============================================================================
class TestConfig:
    max_new_tokens = 512
    temperature = 0.7
    top_p = 0.9
    top_k = 50
    repetition_penalty = 1.1
    do_sample = True
    device = "cuda" if torch.cuda.is_available() else "cpu"
    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')

config = TestConfig()

def cleanup_memory():
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.cuda.synchronize()
    gc.collect()

def print_gpu_memory():
    if torch.cuda.is_available():
        allocated = torch.cuda.memory_allocated() / 1024**3
        reserved = torch.cuda.memory_reserved() / 1024**3
        total = torch.cuda.get_device_properties(0).total_memory / 1024**3
        print(f"💾 GPU: {allocated:.1f}GB allocated, {reserved:.1f}GB reserved, {total:.1f}GB total")

# =============================================================================
# MODEL LOADING
# =============================================================================
print("🚀 COMPLETE SINHALA TESTING SUITE")
print("=" * 60)
print(f"📅 Test Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"🤖 Base Model: {BASE_MODEL_NAME}")
print(f"🎯 LoRA Model: {LORA_MODEL_PATH}")
print(f"📊 Total Test Cases: {len(all_test_cases)}")

# Group tests by category for summary
categories = {}
for test in all_test_cases:
    cat = test.get('category', 'other')
    if cat not in categories:
        categories[cat] = 0
    categories[cat] += 1

print(f"📋 Test Categories:")
for cat, count in categories.items():
    print(f"   {cat.title()}: {count} tests")

# Create output directory
os.makedirs(OUTPUT_DIR, exist_ok=True)
print(f"📁 Output Directory: {OUTPUT_DIR}")

print("\n🔄 Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL_NAME)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
print("✅ Tokenizer loaded")

print("\n🔄 Loading base model with quantization...")
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

base_model = AutoModelForCausalLM.from_pretrained(
    BASE_MODEL_NAME,
    quantization_config=bnb_config,
    device_map="auto",
    torch_dtype=torch.bfloat16,
    trust_remote_code=True
)
print("✅ Base model loaded")

print("\n🔄 Loading LoRA weights...")
try:
    model = PeftModel.from_pretrained(base_model, LORA_MODEL_PATH)
    print("✅ LoRA weights loaded successfully!")
    lora_loaded = True
    model_type = "Fine-tuned LoRA Model"
except Exception as e:
    print(f"❌ Could not load LoRA weights: {e}")
    print("📝 Using base model for comparison...")
    model = base_model
    lora_loaded = False
    model_type = "Base Model (No LoRA)"

model.eval()
print(f"🧠 Using: {model_type}")
print_gpu_memory()

# =============================================================================
# GENERATION FUNCTION
# =============================================================================
def generate_sinhala_response(prompt, max_tokens=512):
    """Generate response using the model"""
    formatted_prompt = f"<start_of_turn>user\n{prompt}<end_of_turn>\n<start_of_turn>model\n"

    inputs = tokenizer(
        formatted_prompt,
        return_tensors="pt",
        truncation=True,
        max_length=1024,
        padding=False
    ).to(config.device)

    start_time = time.time()

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_tokens,
            temperature=config.temperature,
            top_p=config.top_p,
            top_k=config.top_k,
            repetition_penalty=config.repetition_penalty,
            do_sample=config.do_sample,
            pad_token_id=tokenizer.eos_token_id,
            eos_token_id=tokenizer.eos_token_id,
        )

    generation_time = time.time() - start_time

    full_response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    if "<start_of_turn>model\n" in full_response:
        response = full_response.split("<start_of_turn>model\n")[-1].strip()
    else:
        response = full_response.strip()

    return response, generation_time

# =============================================================================
# RESPONSE ANALYSIS FUNCTION
# =============================================================================
def analyze_response(response, test_case):
    """Analyze the quality of response"""
    analysis = {
        "length": len(response),
        "word_count": len(response.split()),
        "contains_sinhala": any(ord(char) >= 0x0D80 and ord(char) <= 0x0DFF for char in response),
        "difficulty": test_case.get("difficulty", "unknown"),
        "category": test_case.get("category", "unknown")
    }

    # Check for expected keywords
    if "expected_keywords" in test_case:
        found_keywords = [kw for kw in test_case["expected_keywords"] if kw in response]
        analysis["keywords_found"] = found_keywords
        analysis["keyword_score"] = len(found_keywords) / len(test_case["expected_keywords"]) if test_case["expected_keywords"] else 0

    # Check for expected answer
    if "expected_answer" in test_case:
        analysis["correct_answer"] = test_case["expected_answer"] in response

    # Special analysis for word generation tasks
    if test_case["id"] == "words_starting_with_a":
        words_with_a = [word for word in response.split() if word.startswith('අ')]
        analysis["words_starting_with_a"] = words_with_a
        analysis["word_count_with_a"] = len(words_with_a)

    # Check for numbers in math problems
    if test_case.get("category") == "math":
        import re
        numbers = re.findall(r'\d+', response)
        analysis["numbers_found"] = numbers

    return analysis

# =============================================================================
# EXECUTE ALL TESTS
# =============================================================================
print("\n🧪 STARTING COMPREHENSIVE TESTING")
print("=" * 60)

test_results = []
category_results = {cat: {"passed": 0, "total": 0} for cat in categories.keys()}
difficulty_results = {"easy": {"passed": 0, "total": 0}, "medium": {"passed": 0, "total": 0}, "advanced": {"passed": 0, "total": 0}}

total_start_time = time.time()

for i, test_case in enumerate(all_test_cases, 1):
    print(f"\n📝 Test {i}/{len(all_test_cases)}: {test_case['id'].upper()}")
    print(f"📊 {test_case['difficulty'].title()} | {test_case['category'].title()}")
    print(f"📄 {test_case['description']}")

    # Show truncated prompt
    prompt_preview = test_case['prompt'][:80] + "..." if len(test_case['prompt']) > 80 else test_case['prompt']
    print(f"❓ {prompt_preview}")

    try:
        # Generate response
        response, gen_time = generate_sinhala_response(test_case['prompt'])

        # Analyze response
        analysis = analyze_response(response, test_case)
        analysis["generation_time"] = gen_time

        # Determine if test passed
        passed = False
        if "keyword_score" in analysis:
            passed = analysis["keyword_score"] >= 0.4  # At least 40% keywords found
        elif "correct_answer" in analysis:
            passed = analysis["correct_answer"]
        elif analysis["contains_sinhala"] and analysis["word_count"] > 3:
            passed = True  # Basic pass for creative tasks

        # Store result
        result = {
            "test_id": test_case["id"],
            "prompt": test_case["prompt"],
            "response": response,
            "analysis": analysis,
            "passed": passed,
            "success": True,
            "timestamp": datetime.now().isoformat()
        }

        test_results.append(result)

        # Update category and difficulty stats
        cat = test_case.get("category", "other")
        diff = test_case.get("difficulty", "unknown")

        if cat in category_results:
            category_results[cat]["total"] += 1
            if passed:
                category_results[cat]["passed"] += 1

        if diff in difficulty_results:
            difficulty_results[diff]["total"] += 1
            if passed:
                difficulty_results[diff]["passed"] += 1

        # Print immediate results
        status = "✅ PASSED" if passed else "⚠️ NEEDS REVIEW"
        print(f"{status} | {gen_time:.2f}s | {analysis['word_count']} words")

        # Show response preview
        preview = response[:150] + "..." if len(response) > 150 else response
        print(f"💬 {preview}")

        # Show specific analysis
        if "keywords_found" in analysis and analysis["keywords_found"]:
            print(f"🎯 Keywords found: {', '.join(analysis['keywords_found'][:3])}")

        if test_case["id"] == "kandy_history_comprehension" and "19" in response:
            print("🎯 CORRECT: Found historical answer!")

        if test_case["id"] == "words_starting_with_a" and analysis.get("word_count_with_a", 0) > 0:
            words = analysis["words_starting_with_a"][:3]
            print(f"🎯 Words with 'අ': {', '.join(words)}")

    except Exception as e:
        print(f"❌ ERROR: {e}")
        result = {
            "test_id": test_case["id"],
            "prompt": test_case["prompt"],
            "response": f"ERROR: {str(e)}",
            "passed": False,
            "success": False,
            "error": str(e),
            "timestamp": datetime.now().isoformat()
        }
        test_results.append(result)

        # Update stats for failed test
        cat = test_case.get("category", "other")
        diff = test_case.get("difficulty", "unknown")
        if cat in category_results:
            category_results[cat]["total"] += 1
        if diff in difficulty_results:
            difficulty_results[diff]["total"] += 1

    # Memory cleanup
    if i % 5 == 0:  # Clean every 5 tests
        cleanup_memory()

total_test_time = time.time() - total_start_time

# =============================================================================
# COMPREHENSIVE RESULTS ANALYSIS
# =============================================================================
print("\n" + "=" * 60)
print("📊 COMPREHENSIVE RESULTS SUMMARY")
print("=" * 60)

successful_tests = [r for r in test_results if r["success"]]
passed_tests = [r for r in test_results if r.get("passed", False)]
failed_tests = [r for r in test_results if not r["success"]]

print(f"✅ Successful Tests: {len(successful_tests)}/{len(test_results)}")
print(f"🎯 Passed Tests: {len(passed_tests)}/{len(test_results)}")
print(f"❌ Failed Tests: {len(failed_tests)}")
print(f"📈 Overall Pass Rate: {len(passed_tests)/len(test_results)*100:.1f}%")
print(f"⏱️ Total Testing Time: {total_test_time:.2f} seconds")

if successful_tests:
    avg_gen_time = sum(r["analysis"]["generation_time"] for r in successful_tests) / len(successful_tests)
    print(f"📈 Average Generation Time: {avg_gen_time:.2f} seconds")

print(f"🧠 Model: {model_type}")
print(f"🎯 LoRA Status: {'✅ Loaded' if lora_loaded else '❌ Not Loaded'}")

# Category-wise results
print(f"\n📋 CATEGORY-WISE RESULTS:")
for cat, stats in category_results.items():
    if stats["total"] > 0:
        pass_rate = stats["passed"] / stats["total"] * 100
        print(f"   {cat.title()}: {stats['passed']}/{stats['total']} ({pass_rate:.1f}%)")

# Difficulty-wise results
print(f"\n🎯 DIFFICULTY-WISE RESULTS:")
for diff, stats in difficulty_results.items():
    if stats["total"] > 0:
        pass_rate = stats["passed"] / stats["total"] * 100
        print(f"   {diff.title()}: {stats['passed']}/{stats['total']} ({pass_rate:.1f}%)")

# =============================================================================
# SAVE RESULTS
# =============================================================================
# Save detailed JSON results
results_file = os.path.join(OUTPUT_DIR, f"complete_test_results_{config.timestamp}.json")
with open(results_file, 'w', encoding='utf-8') as f:
    json.dump({
        "test_info": {
            "base_model": BASE_MODEL_NAME,
            "lora_path": LORA_MODEL_PATH,
            "lora_loaded": lora_loaded,
            "model_type": model_type,
            "test_date": datetime.now().isoformat(),
            "total_tests": len(test_results),
            "successful_tests": len(successful_tests),
            "passed_tests": len(passed_tests),
            "overall_pass_rate": len(passed_tests)/len(test_results)*100,
            "total_time": total_test_time,
            "category_results": category_results,
            "difficulty_results": difficulty_results
        },
        "results": test_results
    }, f, ensure_ascii=False, indent=2)

print(f"💾 Detailed results saved: {results_file}")

# Save human-readable summary
summary_file = os.path.join(OUTPUT_DIR, f"test_summary_{config.timestamp}.txt")
with open(summary_file, 'w', encoding='utf-8') as f:
    f.write("COMPLETE SINHALA GEMMA LORA MODEL TEST SUMMARY\n")
    f.write("=" * 50 + "\n\n")
    f.write(f"Test Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
    f.write(f"Base Model: {BASE_MODEL_NAME}\n")
    f.write(f"LoRA Model: {LORA_MODEL_PATH}\n")
    f.write(f"LoRA Loaded: {'Yes' if lora_loaded else 'No'}\n")
    f.write(f"Training Time: 1.84 hours\n")
    f.write(f"Training Samples: 50,000\n\n")

    f.write(f"OVERALL RESULTS:\n")
    f.write(f"Total Tests: {len(test_results)}\n")
    f.write(f"Passed Tests: {len(passed_tests)}\n")
    f.write(f"Pass Rate: {len(passed_tests)/len(test_results)*100:.1f}%\n\n")

    f.write("CATEGORY BREAKDOWN:\n")
    for cat, stats in category_results.items():
        if stats["total"] > 0:
            pass_rate = stats["passed"] / stats["total"] * 100
            f.write(f"{cat.title()}: {stats['passed']}/{stats['total']} ({pass_rate:.1f}%)\n")

    f.write("\nDIFFICULTY BREAKDOWN:\n")
    for diff, stats in difficulty_results.items():
        if stats["total"] > 0:
            pass_rate = stats["passed"] / stats["total"] * 100
            f.write(f"{diff.title()}: {stats['passed']}/{stats['total']} ({pass_rate:.1f}%)\n")

    f.write("\n" + "="*50 + "\n")
    f.write("INDIVIDUAL TEST RESULTS:\n\n")

    for result in test_results:
        f.write(f"TEST: {result['test_id'].upper()}\n")
        f.write(f"Status: {'✅ PASSED' if result.get('passed') else '⚠️ REVIEW' if result['success'] else '❌ FAILED'}\n")
        if result['success']:
            f.write(f"Generation Time: {result['analysis']['generation_time']:.2f}s\n")
            f.write(f"Word Count: {result['analysis']['word_count']}\n")
        f.write(f"Response: {result['response'][:200]}...\n")
        f.write("-" * 40 + "\n\n")

print(f"📄 Summary saved: {summary_file}")

print(f"\n🎉 COMPLETE TESTING FINISHED!")
print(f"📁 All results saved in: {OUTPUT_DIR}")
print(f"🎯 Overall Performance: {len(passed_tests)/len(test_results)*100:.1f}% pass rate")

# Final cleanup
cleanup_memory()
print_gpu_memory()
print("✅ Complete testing suite finished!")
````

````python
# =============================================================================
# CORRECTED PERPLEXITY CALCULATOR - FOLLOWING STANDARD METHODOLOGY
# Based on the mathematical formula: Perplexity = exp(-1/N * Σ log p(wi|w1...wi-1))
# =============================================================================

import torch
import torch.nn.functional as F
import numpy as np
import os
import json
from datetime import datetime
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    BitsAndBytesConfig
)
from peft import PeftModel
import gc

# =============================================================================
# 1. CONFIGURATION
# =============================================================================
class PerplexityConfig:
    # Model paths - update these to your actual paths
    base_model_name = "google/gemma-3-4b-it"
    lora_model_path = "/content/gemma-sinhala-lora-4b-20250803_1228/final_model"

    # Evaluation settings
    max_length = 512
    batch_size = 1
    device = "cuda" if torch.cuda.is_available() else "cpu"

    # Output settings
    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
    output_dir = f"/content/perplexity_results_{timestamp}"

config = PerplexityConfig()

def cleanup_memory():
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.cuda.synchronize()
    gc.collect()

# =============================================================================
# 2. CORRECTED PERPLEXITY CALCULATION FUNCTION
# =============================================================================
def compute_perplexity_for_batch(model, tokenizer, input_texts, model_name="Model"):
    """
    Compute perplexity following the standard methodology:
    Perplexity = exp(-1/N * Σ log p(wi|w1...wi-1))
    """
    print(f"\n📊 Computing perplexity for {model_name}...")
    print(f"   📝 Processing {len(input_texts)} texts...")

    # Tokenize all texts
    inputs = tokenizer(
        input_texts,
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=config.max_length
    ).to(config.device)

    input_ids = inputs["input_ids"]
    attention_mask = inputs["attention_mask"]

    # Forward pass
    with torch.no_grad():
        outputs = model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits

    # Shift logits and labels for next-token prediction
    # logits[:, :-1, :] -> predictions for positions 0 to N-2
    # input_ids[:, 1:] -> actual tokens at positions 1 to N-1
    shift_logits = logits[:, :-1, :].contiguous()
    shift_labels = input_ids[:, 1:].contiguous()
    shift_attention = attention_mask[:, 1:].contiguous()

    # Convert logits to log probabilities
    log_probs = F.log_softmax(shift_logits, dim=-1)

    # Get log probabilities for the actual next tokens
    # gather() selects the log probability of the correct token at each position
    target_log_probs = log_probs.gather(dim=-1, index=shift_labels.unsqueeze(-1)).squeeze(-1)

    # Apply attention mask to ignore padding tokens
    # Only consider positions where attention_mask = 1
    masked_log_probs = target_log_probs * shift_attention.to(log_probs.dtype)

    # Calculate negative log likelihood for each sequence
    # Sum of log probabilities divided by number of actual tokens (excluding padding)
    sequence_lengths = shift_attention.sum(dim=-1)  # Number of non-padding tokens
    negative_log_likelihood = -masked_log_probs.sum(dim=-1) / sequence_lengths

    # Calculate perplexity: exp(negative_log_likelihood)
    perplexities = torch.exp(negative_log_likelihood)

    # Calculate statistics
    mean_perplexity = torch.mean(perplexities)
    median_perplexity = torch.median(perplexities)

    # Convert to Python lists/floats for JSON serialization
    individual_perplexities = perplexities.cpu().tolist()

    print(f"   ✅ Completed. Mean perplexity: {mean_perplexity.item():.2f}")

    return {
        "individual_perplexities": individual_perplexities,
        "mean_perplexity": mean_perplexity.item(),
        "median_perplexity": median_perplexity.item(),
        "total_sequences": len(input_texts),
        "total_tokens": sequence_lengths.sum().item()
    }

# =============================================================================
# 3. MODEL LOADING
# =============================================================================
print("🧮 CORRECTED PERPLEXITY EVALUATION FOR SINHALA GEMMA")
print("=" * 60)
print(f"📅 Evaluation Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"🤖 Base Model: {config.base_model_name}")
print(f"🎯 LoRA Model: {config.lora_model_path}")

# Create output directory
os.makedirs(config.output_dir, exist_ok=True)

print("\n🔄 Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(config.base_model_name)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

print("🔄 Loading base model...")
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

base_model = AutoModelForCausalLM.from_pretrained(
    config.base_model_name,
    quantization_config=bnb_config,
    device_map="auto",
    torch_dtype=torch.bfloat16,
    trust_remote_code=True
)

print("🔄 Loading LoRA model...")
try:
    lora_model = PeftModel.from_pretrained(base_model, config.lora_model_path)
    print("✅ LoRA model loaded successfully!")
    lora_loaded = True
except Exception as e:
    print(f"❌ Could not load LoRA weights: {e}")
    lora_model = None
    lora_loaded = False

# Set models to evaluation mode
base_model.eval()
if lora_model:
    lora_model.eval()

# =============================================================================
# 4. SINHALA TEST DATA
# =============================================================================
sinhala_test_texts = [
    # Simple sentences (like your document's examples)
    "මම පාසලට යනවා.",
    "අම්මා කුස්සියට කරිය පිසිනවා.",
    "ඔහු පොතක් කියවනවා.",
    "අප්‍රේල් මාසයේ සිංහල අවුරුදු පැමිණේ.",
    "ශ්‍රී ලංකාව දකුණු ආසියාවේ දූපත් රටකි.",

    # Medium complexity
    "මානව පරිණාමය විශ්व ඉතිහාසයේ වැදගත් අවධියකි.",
    "ප්‍රාචීන ලංකාවේ රජවරුන් අනුරාධපුරයේ වාසය කළහ.",
    "විද්‍යාව හා තාක්‍ෂණය සමාජයේ ප්‍රගතියට දායක වේ.",
    "පරිසරය සුරැකීම අපේ හමුදායිත්වයයි.",
    "කලාව හා සාහිත්‍යය සංස්කෘතියේ මූලික අංගයි.",

    # Complex sentences
    "ශ්‍රී ලංකාවේ ඉතිහාසයේ පිහිටුවා ඇති පුරාණ නගරවල සිට නවීන නගර දක්වා සංස්කෘතික පරිවර්තනය සිදුවී ඇත.",
    "අධ්‍යාපන ක්‍ෂේත්‍රයේ ඩිජිටල් තාක්‍ෂණය ප්‍රයෝජනයට ගැනීම නව ඉගෙනුම් ක්‍රමවේද වලට මාර්ගය පෙන්වයි.",

    # Familiar phrases (should have lower perplexity)
    "ගුණ කර ගන්න",
    "සාදරයෙන් පිළිගනිමු",
    "ඔබට ස්තූතියි"
]

print(f"\n📝 Prepared {len(sinhala_test_texts)} Sinhala test sentences")

# =============================================================================
# 5. CALCULATE PERPLEXITY FOR BOTH MODELS
# =============================================================================
results = {}

# Base model perplexity
print(f"\n🔄 Evaluating Base Gemma Model...")
base_results = compute_perplexity_for_batch(
    base_model, tokenizer, sinhala_test_texts, "Base Gemma Model"
)
results["base_model"] = base_results

# LoRA model perplexity (if available)
if lora_model:
    print(f"\n🔄 Evaluating LoRA Fine-tuned Model...")
    lora_results = compute_perplexity_for_batch(
        lora_model, tokenizer, sinhala_test_texts, "LoRA Fine-tuned Model"
    )
    results["lora_model"] = lora_results

# =============================================================================
# 6. DETAILED ANALYSIS AND COMPARISON
# =============================================================================
print(f"\n📈 DETAILED PERPLEXITY ANALYSIS")
print("=" * 50)

# Print individual sentence results
if "lora_model" in results:
    print(f"\n📝 INDIVIDUAL SENTENCE PERPLEXITY (LoRA Model):")
    print("-" * 60)
    for i, (text, ppl) in enumerate(zip(sinhala_test_texts, results["lora_model"]["individual_perplexities"])):
        text_preview = text[:40] + "..." if len(text) > 40 else text
        print(f"{i+1:2d}. {ppl:6.2f} | {text_preview}")

# Comparative analysis
if "base_model" in results and "lora_model" in results:
    base_ppl = results["base_model"]["mean_perplexity"]
    lora_ppl = results["lora_model"]["mean_perplexity"]
    improvement = ((base_ppl - lora_ppl) / base_ppl) * 100

    print(f"\n🆚 MODEL COMPARISON:")
    print("-" * 30)
    print(f"📊 Base Model Mean Perplexity:  {base_ppl:.2f}")
    print(f"🎯 LoRA Model Mean Perplexity:  {lora_ppl:.2f}")
    print(f"📈 Improvement: {improvement:+.1f}%")

    if lora_ppl < base_ppl:
        print(f"✅ SUCCESS: LoRA model shows BETTER performance!")
        print(f"🎉 Your fine-tuning was effective!")
    else:
        print(f"⚠️ LoRA model shows higher perplexity than base model")

    # Statistical analysis
    base_individual = results["base_model"]["individual_perplexities"]
    lora_individual = results["lora_model"]["individual_perplexities"]

    print(f"\n📊 STATISTICAL SUMMARY:")
    print(f"   Base Model  - Mean: {np.mean(base_individual):.2f}, Std: {np.std(base_individual):.2f}")
    print(f"   LoRA Model  - Mean: {np.mean(lora_individual):.2f}, Std: {np.std(lora_individual):.2f}")

# Performance interpretation
if "lora_model" in results:
    lora_ppl = results["lora_model"]["mean_perplexity"]

    print(f"\n🎯 PERFORMANCE INTERPRETATION:")
    print("-" * 35)

    if lora_ppl < 10:
        quality = "Excellent 🌟"
        interpretation = "Outstanding performance for Sinhala generation"
    elif lora_ppl < 20:
        quality = "Very Good ✅"
        interpretation = "Strong performance, suitable for production use"
    elif lora_ppl < 50:
        quality = "Good 👍"
        interpretation = "Solid performance, effective fine-tuning"
    elif lora_ppl < 100:
        quality = "Fair 📊"
        interpretation = "Reasonable performance, room for improvement"
    else:
        quality = "Needs Improvement ⚠️"
        interpretation = "Consider additional training or different approach"

    print(f"   Quality Level: {quality}")
    print(f"   Interpretation: {interpretation}")

# =============================================================================
# 7. SAVE RESULTS (SAME AS BEFORE)
# =============================================================================
print(f"\n💾 SAVING RESULTS...")

# Save detailed JSON results
results_file = os.path.join(config.output_dir, f"corrected_perplexity_results_{config.timestamp}.json")

evaluation_info = {
    "evaluation_info": {
        "methodology": "Standard perplexity calculation: exp(-1/N * Σ log p(wi|w1...wi-1))",
        "timestamp": datetime.now().isoformat(),
        "base_model": config.base_model_name,
        "lora_model_path": config.lora_model_path,
        "lora_loaded": lora_loaded,
        "test_texts_count": len(sinhala_test_texts),
        "max_length": config.max_length
    },
    "results": results,
    "test_texts": sinhala_test_texts
}

with open(results_file, 'w', encoding='utf-8') as f:
    json.dump(evaluation_info, f, ensure_ascii=False, indent=2)

print(f"✅ Results saved: {results_file}")

# =============================================================================
# 8. FINAL SUMMARY
# =============================================================================
print(f"\n🎉 CORRECTED PERPLEXITY EVALUATION COMPLETED!")
print("=" * 55)

if "lora_model" in results:
    lora_ppl = results["lora_model"]["mean_perplexity"]
    print(f"🎯 Your LoRA Model Mean Perplexity: {lora_ppl:.2f}")

    if "base_model" in results:
        improvement = ((results["base_model"]["mean_perplexity"] - lora_ppl) / results["base_model"]["mean_perplexity"]) * 100
        print(f"📈 Improvement over base model: {improvement:+.1f}%")

print(f"📁 Results saved in: {config.output_dir}")
print("✅ Evaluation completed using standard methodology!")

cleanup_memory()
````

````python
# =============================================================================
# DEBUGGED PERPLEXITY CALCULATOR WITH DETAILED DIAGNOSTICS
# Fixes the issues identified in your previous results
# =============================================================================

import torch
import torch.nn.functional as F
import numpy as np
import os
import json
from datetime import datetime
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    BitsAndBytesConfig
)
from peft import PeftModel
import gc

# =============================================================================
# 1. ENHANCED CONFIGURATION WITH DEBUGGING
# =============================================================================
class PerplexityConfig:
    base_model_name = "google/gemma-3-4b-it"
    lora_model_path = "/content/gemma-sinhala-lora-4b-20250803_1228/final_model"

    # More conservative settings for debugging
    max_length = 256  # Reduced from 512
    batch_size = 1    # Process one at a time for debugging
    device = "cuda" if torch.cuda.is_available() else "cpu"

    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
    output_dir = f"/content/debugged_perplexity_{timestamp}"

config = PerplexityConfig()

def cleanup_memory():
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.cuda.synchronize()
    gc.collect()

def debug_tokenization(tokenizer, text):
    """Debug tokenization to identify issues"""
    print(f"   🔍 Debug: '{text[:30]}...'")

    # Test tokenization
    tokens = tokenizer.tokenize(text)
    token_ids = tokenizer.encode(text, return_tensors="pt")

    print(f"      📝 Text length: {len(text)} chars")
    print(f"      🔢 Token count: {len(tokens)}")
    print(f"      📊 Token IDs shape: {token_ids.shape}")
    print(f"      🎯 First 5 tokens: {tokens[:5] if len(tokens) > 5 else tokens}")

    # Check for unknown tokens
    unk_token_id = tokenizer.unk_token_id if hasattr(tokenizer, 'unk_token_id') else None
    if unk_token_id:
        unk_count = (token_ids == unk_token_id).sum().item()
        print(f"      ⚠️ Unknown tokens: {unk_count}")

    return len(tokens), token_ids.shape[1]

# =============================================================================
# 2. FIXED PERPLEXITY CALCULATION WITH DEBUGGING
# =============================================================================
def compute_perplexity_single_text(model, tokenizer, text, model_name="Model"):
    """
    Compute perplexity for a single text with detailed debugging
    """
    try:
        # Tokenize with detailed logging
        inputs = tokenizer(
            text,
            return_tensors="pt",
            padding=False,  # No padding for single text
            truncation=True,
            max_length=config.max_length,
            add_special_tokens=True
        ).to(config.device)

        input_ids = inputs["input_ids"]
        seq_len = input_ids.size(1)

        print(f"      📊 Sequence length: {seq_len}")

        if seq_len < 2:
            print(f"      ⚠️ Sequence too short, skipping")
            return None, "Too short"

        # Forward pass with error handling
        with torch.no_grad():
            outputs = model(input_ids)
            logits = outputs.logits

        print(f"      🧮 Logits shape: {logits.shape}")

        # Prepare for next-token prediction
        shift_logits = logits[:, :-1, :].contiguous()  # Remove last position
        shift_labels = input_ids[:, 1:].contiguous()   # Remove first position

        # Flatten for cross-entropy calculation
        flat_logits = shift_logits.view(-1, shift_logits.size(-1))
        flat_labels = shift_labels.view(-1)

        # Calculate cross-entropy loss (negative log likelihood)
        loss = F.cross_entropy(flat_logits, flat_labels, reduction='mean')

        # Convert to perplexity
        perplexity = torch.exp(loss).item()

        print(f"      📈 Loss: {loss.item():.4f}")
        print(f"      🎯 Perplexity: {perplexity:.2f}")

        # Sanity checks
        if perplexity > 10000:
            print(f"      ⚠️ WARNING: Very high perplexity detected!")
        elif perplexity < 1:
            print(f"      ⚠️ WARNING: Perplexity below 1 (impossible)")

        return perplexity, "Success"

    except Exception as e:
        print(f"      ❌ Error: {str(e)}")
        return None, str(e)

def compute_perplexity_for_model(model, tokenizer, texts, model_name="Model"):
    """
    Compute perplexity for all texts with detailed debugging
    """
    print(f"\n🔍 DEBUGGING {model_name.upper()}")
    print("-" * 50)

    # Test model state
    print(f"   🧠 Model training mode: {model.training}")
    print(f"   💾 Model device: {next(model.parameters()).device}")
    print(f"   🎯 Model dtype: {next(model.parameters()).dtype}")

    # Set model to eval mode explicitly
    model.eval()
    print(f"   ✅ Model set to eval mode")

    valid_perplexities = []
    detailed_results = []

    for i, text in enumerate(texts):
        print(f"\n   📝 Text {i+1}/{len(texts)}:")

        # Debug tokenization
        token_count, seq_len = debug_tokenization(tokenizer, text)

        # Calculate perplexity
        perplexity, status = compute_perplexity_single_text(model, tokenizer, text, model_name)

        result = {
            "text": text,
            "text_length": len(text),
            "token_count": token_count,
            "perplexity": perplexity,
            "status": status
        }

        detailed_results.append(result)

        if perplexity is not None and perplexity > 0:
            valid_perplexities.append(perplexity)

        # Memory cleanup after each text
        cleanup_memory()

    # Calculate statistics
    if valid_perplexities:
        mean_ppl = np.mean(valid_perplexities)
        median_ppl = np.median(valid_perplexities)
        std_ppl = np.std(valid_perplexities)

        print(f"\n   📊 {model_name} Results:")
        print(f"      Valid texts: {len(valid_perplexities)}/{len(texts)}")
        print(f"      Mean perplexity: {mean_ppl:.2f}")
        print(f"      Median perplexity: {median_ppl:.2f}")
        print(f"      Std deviation: {std_ppl:.2f}")
        print(f"      Min perplexity: {min(valid_perplexities):.2f}")
        print(f"      Max perplexity: {max(valid_perplexities):.2f}")

        return {
            "individual_perplexities": valid_perplexities,
            "mean_perplexity": mean_ppl,
            "median_perplexity": median_ppl,
            "std_perplexity": std_ppl,
            "valid_count": len(valid_perplexities),
            "total_count": len(texts),
            "detailed_results": detailed_results
        }
    else:
        print(f"   ❌ No valid perplexity scores calculated!")
        return None

# =============================================================================
# 3. MODEL LOADING WITH ENHANCED DEBUGGING
# =============================================================================
print("🚨 DEBUGGED PERPLEXITY EVALUATION FOR SINHALA GEMMA")
print("=" * 60)
print(f"📅 Evaluation Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

os.makedirs(config.output_dir, exist_ok=True)

print("\n🔄 Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(config.base_model_name)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# Debug tokenizer info
print(f"   📝 Tokenizer vocab size: {tokenizer.vocab_size}")
print(f"   🎯 Pad token: {tokenizer.pad_token}")
print(f"   🎯 EOS token: {tokenizer.eos_token}")

print("\n🔄 Loading base model...")
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

base_model = AutoModelForCausalLM.from_pretrained(
    config.base_model_name,
    quantization_config=bnb_config,
    device_map="auto",
    torch_dtype=torch.bfloat16,
    trust_remote_code=True
)

print("\n🔄 Loading LoRA model...")
try:
    lora_model = PeftModel.from_pretrained(base_model, config.lora_model_path)
    print("✅ LoRA model loaded successfully!")
    lora_loaded = True

    # Critical: Create separate instances to avoid identical results
    print("🔄 Creating separate model instances...")

    # Base model for comparison (copy)
    base_model_eval = AutoModelForCausalLM.from_pretrained(
        config.base_model_name,
        quantization_config=bnb_config,
        device_map="auto",
        torch_dtype=torch.bfloat16,
        trust_remote_code=True
    )
    base_model_eval.eval()

    # LoRA model
    lora_model.eval()

    print("✅ Separate model instances created")

except Exception as e:
    print(f"❌ Could not load LoRA weights: {e}")
    lora_model = None
    base_model_eval = base_model
    lora_loaded = False

# =============================================================================
# 4. SIMPLIFIED TEST DATA FOR DEBUGGING
# =============================================================================
# Start with simpler, shorter texts for debugging
simple_sinhala_texts = [
    "මම යනවා.",           # Very simple
    "ඔහු කනවා.",          # Very simple
    "අම්මා පිසිනවා.",       # Simple
    "පාසල හොඳයි.",         # Simple
    "ලංකාව දූපතකි.",        # Simple
]

print(f"\n📝 Testing with {len(simple_sinhala_texts)} simple Sinhala sentences")

# =============================================================================
# 5. RUN DEBUGGED EVALUATION
# =============================================================================
results = {}

# Test base model
print(f"\n🔍 TESTING BASE MODEL:")
base_results = compute_perplexity_for_model(
    base_model_eval, tokenizer, simple_sinhala_texts, "Base Gemma Model"
)

if base_results:
    results["base_model"] = base_results

cleanup_memory()

# Test LoRA model
if lora_model and lora_loaded:
    print(f"\n🔍 TESTING LORA MODEL:")
    lora_results = compute_perplexity_for_model(
        lora_model, tokenizer, simple_sinhala_texts, "LoRA Fine-tuned Model"
    )

    if lora_results:
        results["lora_model"] = lora_results

# =============================================================================
# 6. COMPARISON AND ANALYSIS
# =============================================================================
print(f"\n📊 COMPARISON ANALYSIS")
print("=" * 40)

if "base_model" in results and "lora_model" in results:
    base_ppl = results["base_model"]["mean_perplexity"]
    lora_ppl = results["lora_model"]["mean_perplexity"]

    print(f"📊 Base Model Mean: {base_ppl:.2f}")
    print(f"🎯 LoRA Model Mean: {lora_ppl:.2f}")

    if abs(base_ppl - lora_ppl) < 0.01:
        print(f"⚠️ WARNING: Models show identical performance!")
        print(f"   This suggests the LoRA weights aren't being applied correctly")
    else:
        improvement = ((base_ppl - lora_ppl) / base_ppl) * 100
        print(f"📈 Improvement: {improvement:+.1f}%")

        if lora_ppl < base_ppl:
            print(f"✅ LoRA model performs better!")
        else:
            print(f"⚠️ Base model performs better")

elif "base_model" in results:
    print(f"📊 Base Model Mean: {results['base_model']['mean_perplexity']:.2f}")
    print(f"❌ LoRA model evaluation failed")

elif "lora_model" in results:
    print(f"🎯 LoRA Model Mean: {results['lora_model']['mean_perplexity']:.2f}")
    print(f"❌ Base model evaluation failed")

else:
    print(f"❌ Both evaluations failed!")

# =============================================================================
# 7. SAVE DEBUGGING RESULTS
# =============================================================================
print(f"\n💾 SAVING DEBUGGING RESULTS...")

debug_file = os.path.join(config.output_dir, f"debug_perplexity_{config.timestamp}.json")

debug_info = {
    "debug_info": {
        "timestamp": datetime.now().isoformat(),
        "base_model": config.base_model_name,
        "lora_model_path": config.lora_model_path,
        "lora_loaded": lora_loaded,
        "test_texts": simple_sinhala_texts,
        "issues_identified": [
            "Previous identical scores for both models",
            "Extremely high perplexity values",
            "Possible model state confusion"
        ],
        "fixes_applied": [
            "Separate model instances",
            "Explicit eval() mode setting",
            "Single-text processing with debugging",
            "Enhanced error handling"
        ]
    },
    "results": results
}

with open(debug_file, 'w', encoding='utf-8') as f:
    json.dump(debug_info, f, ensure_ascii=False, indent=2)

print(f"✅ Debug results saved: {debug_file}")

print(f"\n🎉 DEBUGGING COMPLETED!")
print("✅ Check the results above to see if the issues are resolved")

cleanup_memory()
````
